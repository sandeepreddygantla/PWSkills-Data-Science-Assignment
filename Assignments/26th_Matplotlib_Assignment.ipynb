# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

**Matplotlib** is a popular plotting library in Python used for creating a variety of high-quality, publication-ready 2D plots, charts, and graphs. It allows users to create a wide range of static, animated, and interactive visualizations in Python.

Matplotlib is used in various domains such as scientific computing, data visualization, machine learning, and data analysis. It is widely used in academic research, business, and industry for visualizing data and communicating insights.

Here are five common types of plots that can be created using the Pyplot module of Matplotlib:

**Line plot:** Shows the relationship between two variables on a continuous scale with a line connecting the data points.

**Scatter plot:** Displays the relationship between two variables as a set of points on a 2D plane.

**Bar plot:** Compares the values of different categories by using bars of varying heights or lengths.

**Histogram:** Displays the distribution of a single variable by dividing it into a set of intervals and counting the number of observations that fall into each interval.

**Pie chart:** Displays the relative proportions of categories in a dataset as slices of a circular pie.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A **scatter plot** is a type of plot used to display the relationship between two variables. It is a two-dimensional plot that uses dots to represent the values of the two variables. One variable is plotted on the x-axis and the other variable is plotted on the y-axis.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(9,4))
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.scatter(x,y, c='r')
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

The **subplot()** function in Matplotlib is used to create multiple plots within a single figure. It allows the user to arrange subplots in a grid of rows and columns, where each cell of the grid contains a single plot.

The **subplot()** function takes three arguments: the number of rows, the number of columns, and the index of the subplot. The index starts from 1 and increases from left to right and top to bottom. This allows the user to specify the position of each subplot within the grid.

```python
# line 1
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
#line 2
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
#line 3
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([10, 20, 30, 40, 50, 60])
#line 4
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot()
ax.plot(x,y,c='r')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is First Plot")
plt.show()

fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot()
ax.plot(x1,y2,c='b')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is second Plot")
plt.show()

fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot()
ax.plot(x2,y2,c='g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is third Plot")
plt.show()

fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot()
ax.plot(x3,y3,c='y')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is fourth Plot")
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A **bar plot** is a graphical representation of data that displays categorical data using rectangular bars. The height or length of each bar corresponds to the frequency, count, or percentage of the data within that category.

**Bar plots** are commonly used to compare the relative sizes of different categories or to visualize changes in data over time. They are particularly useful for showing differences between discrete categories or groups of data, such as different age groups, income levels, or product sales.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
fig = plt.figure(figsize=(10,3))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is Bar Plot")
plt.bar(company,profit)
plt.show()
```

```python
fig = plt.figure(figsize=(10,3))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is Horizontal Bar Plot")
plt.barh(company,profit)
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A **box plot,** also known as a box-and-whisker plot, is a graphical representation of data that displays the distribution of a dataset. It is based on a five-number summary, which includes the minimum value, lower quartile (25th percentile), median (50th percentile), upper quartile (75th percentile), and maximum value.

A **box plot** consists of a box, which spans the interquartile range (IQR) and shows the middle 50% of the data. The median is represented by a line within the box. Whiskers extend from the box to the minimum and maximum values, excluding any outliers, which are plotted as individual points.

**Box plots** are used to identify the spread and skewness of a dataset and to compare the distributions of multiple datasets. They are particularly useful for displaying data with many outliers or non-normal distributions. Box plots can also be used to detect any asymmetry or skewness in the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
a = [box1,box2]
fig = plt.figure(figsize=(10,3))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is Box Plot")
plt.boxplot(a)
plt.show()
```
