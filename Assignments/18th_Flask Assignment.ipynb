# Q1. What is Flask Framework? What are the advantages of Flask Framework?

**Flask** is a lightweight and flexible web framework for building web applications in Python. It provides a simple and intuitive API for creating web applications, with features such as routing, templating, and request handling built in. Flask is designed to be easy to use and learn, and it allows developers to create web applications quickly and with minimal boilerplate code.

Here are some of the key advantages of Flask framework:

**Lightweight:** Flask is a lightweight framework with a small codebase, making it fast and efficient.

**Flexibility:** Flask is a flexible framework that allows developers to customize and extend it to meet the specific needs of their application.

**Easy to learn:** Flask is designed to be easy to learn and use, making it accessible to developers of all skill levels.

**Large community:** Flask has a large and active community of developers, which means that there are many third-party packages and extensions available to add new features and functionality to your web application.

**Integrated with Python:** Flask is built on top of Python, which is a popular and widely-used programming language. This means that developers can leverage the full power and flexibility of Python in their web applications.

**Quick development:** With Flask, developers can create web applications quickly and with minimal boilerplate code, which is ideal for prototyping and rapid application development.

**Scalability:** Flask can be used to build scalable web applications, with support for caching, load balancing, and other features that make it easy to handle high levels of traffic.

# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

# Q3. What is App routing in Flask? Why do we use app routes?

**App routing** is a mechanism provided by the Flask web framework that allows developers to define URLs (routes) that correspond to specific pages or functions in their web application. In Flask, app routes are defined using a Python decorator called @app.route().

**App routes** are an essential part of any Flask application because they provide a way to map URLs to specific functions or pages in the application. This allows users to navigate through the application and perform specific actions, such as viewing a page, submitting a form, or logging in.

There are several reasons why app routes are important in Flask applications:

**URL structure:** App routes provide a way to define the structure of the URLs in your application, making it easy for users to navigate and understand the content of your site.

**Request handling:** App routes allow developers to handle specific types of HTTP requests (such as GET, POST, PUT, and DELETE) and perform different actions based on the request method.

**View functions:** App routes provide a way to map URLs to specific view functions in your application, allowing you to separate your application logic into modular functions that can be easily tested and maintained.

**User interaction:** App routes enable users to interact with your application by submitting forms, logging in, and performing other actions that are essential to the functionality of the site.

# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation

Location: India

Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<p>Company Name: ABC Corporation</p> <p>Location: India</p><p> Contact Detail: 999-999-9999 </p>"

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<p>Company Name: ABC Corporation</p> <p>Location: India</p><p> Contact Detail: 999-999-9999 </p>"

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

with app.test_request_context():
    print(url_for('welcome'))

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
